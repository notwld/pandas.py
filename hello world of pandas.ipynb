# One Hawt Encoding : Cleaning data 
for example replacing null values with mean values

```python
data.isnull().sum()
```

```python
value = int(data['user rating score'].mean())
value
```

```python
data.replace(to_replace=np.nan,value=value,inplace=True) #implace does permenant change 
```
